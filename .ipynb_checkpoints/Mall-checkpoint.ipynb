```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("Mall_Customers.csv")
print(df.head())
```

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
import os
import warnings

# Fix MKL + Windows warning
os.environ["OMP_NUM_THREADS"] = "1"
warnings.filterwarnings("ignore", category=UserWarning)

# Step 4: Elbow Method to find optimal clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
df['Cluster'] = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)',
                hue='Cluster', palette='Set1', data=df, s=100)
plt.title('Customer Segmentation')
plt.show()
```

```python
print(df.groupby('Cluster').mean(numeric_only=True))
```

```python
cluster_summary = df.groupby('Cluster').agg({
    'Age': 'mean',
    'Annual Income (k$)': 'mean',
    'Spending Score (1-100)': 'mean'
}).round(2)

cluster_summary.rename(columns={
    'Age': 'Avg Age',
    'Annual Income (k$)': 'Avg Annual Income (k$)',
    'Spending Score (1-100)': 'Avg Spending Score'
}, inplace=True)

print(cluster_summary)
```

```python
cluster_summary.to_csv("cluster_summary.csv")
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x='Annual Income (k$)', 
    y='Spending Score (1-100)',
    hue='Cluster', 
    palette='Set1', 
    data=df, 
    s=100
)
```

```python
centers = kmeans.cluster_centers_
plt.scatter(
    centers[:, 0], centers[:, 1],
    c='black', s=300, alpha=0.6, marker='X', label='Centroids'
)

plt.title('Customer Segmentation', fontsize=16)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
